```python
import pandas as pd
df = pd.read_csv('130001_tokyo_covid19_patients_per_report_date.csv')
```

```python
df = df[['公表_年月日', '日別陽性者数']]
```

```python
df.rename(columns={'公表_年月日': 'Dates', '日別陽性者数': 'The_positive_cases'}, inplace=True)
```

```python
df
```

```python
df['Dates'] = pd.to_datetime(df['Dates'])
```

```python
df_monthly = df.resample('M', on='Dates').sum()
df_monthly['Dates'] = df_monthly.index.strftime('%Y-%m-01')
df_monthly = df_monthly[['Dates', 'The_positive_cases']]
df_monthly
```

```python
df_monthly['Dates'] = df_monthly['Dates'].str[:-1]
#df_monthly['Dates'] = pd.to_datetime(df_monthly['Dates']).dt.to_period('M')
```
