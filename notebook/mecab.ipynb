```python
import pandas as pd
import MeCab
from collections import Counter
wakati = MeCab.Tagger("-Owakati")

def count_all_nouns(text):
    all_nouns = []
    node = wakati.parseToNode(text)
    while node:
        if node.feature.split(',')[0] == '名詞':  
            all_nouns.append(node.surface)
        node = node.next
    return Counter(all_nouns)

print(df.dtypes)

texts = df['Extracted'].fillna('').astype(str).tolist()
plus_texts = df['Plus'].fillna('').astype(str).tolist()

all_text = ' '.join(texts + plus_texts)

all_nouns_counts = count_all_nouns(all_text)

sorted_nouns_counts = sorted(all_nouns_counts.items(), key=lambda x: x[1], reverse=True)
for noun, count in sorted_nouns_counts:
    print(f"{noun}: {count}")
```

```python
keywords = ['茂木', '副総裁|麻生', '森山', '山口', '那津男','世耕','遠藤','根本','磯崎','関口','松野','木原','幹雄','萩生田','甘利','雅和','武藤','国場','中本','高木','松山','藤川','堀内','野上','関根']

results = []
for keyword in keywords:
    if '|' in keyword:
        keywords_split = keyword.split('|')
        condition_extracted = df['Extracted'].fillna('').astype(str).str.contains(keywords_split[0], regex=False)
        for key in keywords_split[1:]:
            condition_extracted |= df['Extracted'].fillna('').astype(str).str.contains(key, regex=False)
        df_extracted = df[condition_extracted]
        
        condition_plus = df['Plus'].fillna('').astype(str).str.contains(keywords_split[0], regex=False)
        for key in keywords_split[1:]:
            condition_plus |= df['Plus'].fillna('').astype(str).str.contains(key, regex=False)
        df_plus = df[condition_plus]
    else:
        if keyword in ['山口', '那津男']:
            condition_extracted = df['Extracted'].fillna('').astype(str).str.match(keyword)
            condition_plus = df['Plus'].fillna('').astype(str).str.match(keyword)
        else:
            condition_extracted = df['Extracted'].fillna('').astype(str).str.contains(keyword, regex=False)
            condition_plus = df['Plus'].fillna('').astype(str).str.contains(keyword, regex=False)
        
        df_extracted = df[condition_extracted]
        df_plus = df[condition_plus]
    
    count_extracted = len(df_extracted)
    count_plus = len(df_plus)
    total_count = count_extracted + count_plus
    
    results.append((keyword, total_count))

results.sort(key=lambda x: x[1], reverse=True)

for keyword, total_count in results:
    print(f"Keyword: {keyword}")
    print(f"Total rows: {total_count}")
    print()
```
