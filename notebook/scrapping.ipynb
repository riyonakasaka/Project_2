```python
#get the elements from all urls

import requests
from bs4 import BeautifulSoup

large_url_list = []
for i in range(1, 47):
    large_url_list.append(f"https://www.nikkei.com/topics/22041100?page={i}")

all_mini_url_list = []
for url in large_url_list:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    for a in soup.find('div', class_='contents_c1erz3ur').find_all('a', class_='fauxBlockLink_f6t5v6i'):
        url_piece = a['href']
        all_mini_url_list.append(f"https://www.nikkei.com{url_piece}")

#  get the elements from <p class='paragraph_p18mfke4'> and <time datetime> 
all_paragraphs = []

for mini_url in all_mini_url_list:
    response = requests.get(mini_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    # <time datetime> 
    time_element = soup.find('time')
    if time_element and 'datetime' in time_element.attrs:
        date = time_element['datetime'][:10]  
    else:
        date = "No date found"
    all_paragraphs.append(f"{date}:")
    # <p class='paragraph_p18mfke4'> 
    paragraphs = soup.find_all('p', class_='paragraph_p18mfke4')
    for paragraph in paragraphs:
        text = paragraph.text.replace("", "")  
        all_paragraphs.append(text)
    all_paragraphs.append('------')

for para in all_paragraphs:
    print(para)
```

```python
#to put into the df

import requests
from bs4 import BeautifulSoup
import pandas as pd

large_url_list = []
for i in range(1, 47):
    large_url_list.append(f"https://www.nikkei.com/topics/22041100?page={i}")

all_mini_url_list = []

for url in large_url_list:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    for a in soup.find('div', class_='contents_c1erz3ur').find_all('a', class_='fauxBlockLink_f6t5v6i'):
        url_piece = a['href']
        all_mini_url_list.append(f"https://www.nikkei.com{url_piece}")

data = []
for mini_url in all_mini_url_list:
    response = requests.get(mini_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    time_element = soup.find('time')
    if time_element and 'datetime' in time_element.attrs:
        date = time_element['datetime'][:10]
    else:
        date = "No date found"
    paragraphs = soup.find_all('p', class_='paragraph_p18mfke4')
    
    for paragraph in paragraphs:
        text = paragraph.text.replace("", "") 
        data.append({'Date': date, 'Text': text})

df = pd.DataFrame(data)
print(df.head())
```

```python
def extract_paragraph(text):
    matches = re.search(r'[][^]*', text)
    if matches:
        return matches.group(0)[:-2].strip()  
    else:
        return ''
filtered_df = df[df['Paragraph'].str.contains('')]  # filtering''
filtered_df['Extracted'] = filtered_df['Paragraph'].apply(extract_paragraph)

filtered_df['Extracted'] = filtered_df['Extracted'].str.replace('', '')
filtered_df['P']
```

```python
def extract_phrases(text):
    matches = re.search(r'([^]*)', text)
    if matches:
        return matches.group(1)
    else:
        return ''
filtered_df['Plus'] = filtered_df['Paragraph'].apply(lambda x: extract_phrases(str(x)))
filtered_df.to_csv('filtered.csv')
```

```python
def extract_previous_sentence(filtered_df):
    results = []
    
    for index, row in filtered_df.iterrows():
        extracted_text = row['Extracted']
        if '' not in extracted_text:
            matches = re.search(r'([][^]+[][^]+[][^]+)[][^]+$', row['Paragraph'])
            if matches:
                previous_sentence = matches.group(1).strip()
            else:
                previous_sentence = ''
            results.append(previous_sentence)
        else:
            results.append('')
    return results
filtered_df['Plus'] = extract_previous_sentence(filtered_df)
filtered_df['Place'] = filtered_df.apply(lambda row: row['Paragraph'] if '' not in row['Extracted'] else '', axis=1)
filtered_df[['Date', 'Paragraph', 'Extracted', 'Plus', 'Place']].to_excel('kaisyoku.xlsx')
```
