```python
import pandas as pd
```

```python
df = pd.read_csv('filtered.csv')
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df_date_paragraph = df[['Date', 'Paragraph']]
df_date_paragraph
```

```python
filtered_rows = []
remaining_rows = []

for index, row in df_date_paragraph.iterrows():
    if '11時' in row['Paragraph'] or '12時' in row['Paragraph'] or '13時' in row['Paragraph']:
        filtered_rows.append(row)
    else:
        remaining_rows.append(row)

filtered_df = pd.DataFrame(filtered_rows)
remaining_df = pd.DataFrame(remaining_rows)

len(filtered_df)
```

```python
filtered_df['Date'] = pd.to_datetime(filtered_df['Date'], errors='coerce')
filtered_df = filtered_df.dropna(subset=['Date'])

filtered_df['YearMonth'] = filtered_df['Date'].dt.to_period('M')

monthly_counts = filtered_df.groupby('YearMonth').size().reset_index(name='Lunch')
monthly_counts
```

```python
remaining_df['Date'] = pd.to_datetime(remaining_df['Date'], errors='coerce')
remaining_df = remaining_df.dropna(subset=['Date'])

remaining_df['YearMonth'] = remaining_df['Date'].dt.to_period('M')
monthly_counts = remaining_df.groupby('YearMonth').size().reset_index(name='Dinner')
monthly_counts
```

```python

filtered_df['Date'] = pd.to_datetime(filtered_df['Date'], errors='coerce')
filtered_df = filtered_df.dropna(subset=['Date'])
filtered_df['YearMonth'] = filtered_df['Date'].dt.to_period('M')

remaining_df['Date'] = pd.to_datetime(remaining_df['Date'], errors='coerce')
remaining_df = remaining_df.dropna(subset=['Date'])
remaining_df['YearMonth'] = remaining_df['Date'].dt.to_period('M')

monthly_counts_filtered = filtered_df.groupby('YearMonth').size().reset_index(name='Lunch')
monthly_counts_remaining = remaining_df.groupby('YearMonth').size().reset_index(name='Dinner')

merged_counts = pd.merge(monthly_counts_filtered, monthly_counts_remaining, on='YearMonth', how='outer')

df= merged_counts
```

```python
df_filled = df.fillna(0)

df_filled['Lunch'] = df_filled['Lunch'].round(0).astype(int)
df_filled['Dinner'] = df_filled['Dinner'].round(0).astype(int)
```

```python
df_filled.to_csv('lunchdinner.csv')
```
